```python
import numpy as np
import matplotlib.pyplot as plt

w=3
b=0.5
x_lin = np.linspace(0,100,101)
y=(x_lin+np.random.randn(101)*5)*w+b
plt.plot(x_lin,y,'b.', label = 'data points')
plt.title('Thats the data point')
plt.legend(loc=2)
plt.show()

y_hat = w*x_lin+b
plt.plot(x_lin,y,'b.',label = 'data points')
plt.plot(x_lin,y_hat,'r-', label = 'Prediction')
plt.legend(loc=2)
plt.show()
def MAE(y,y_hat):
    mae = sum(abs(y-y_hat))/len(y)
    print("the MAE is %.3f"%mae)
    return mae
def MSE(y,y_hat):
    mse = sum(pow(y-y_hat,2))/len(y)
    print("the MSE is %.3f"%mse)
    return mse
```
